```python
import pandas as pd

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
data = pd.read_csv('NBA_Season_2020.csv')
```

```python
data.to_csv('NBA_Season_2020.csv')
```

```python
data.head()
```

```python
most_popular_words = {}

for idx, posession in enumerate(data['Posession'][:20000]):
    #print(posession)
    all_lowercase = [word.lower() for word in posession.split() if '.' not in word.lower() 
                                                                 and word.lower not in stopwords.words('english')
                                                                 and word.isnumeric() == False]
    distinct_words = list(set(all_lowercase))
    #print(distinct_words)
    
    for word in distinct_words:
        if word in most_popular_words:
            most_popular_words[word] += 1
        else:
            most_popular_words[word] = 1
    
    if idx % 500 == 0:
        print(idx, ' rows Completed')
```

```python
most_popular_words = sorted(most_popular_words.items(), key=lambda x: x[1], reverse=True)

for i in most_popular_words:
    print(i[0], i[1])
```
