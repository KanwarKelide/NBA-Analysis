```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
conn = sqlite3.connect('NBA.db')
```

```python
foul_info = pd.read_sql('SELECT * FROM foul_information WHERE IsStarter = 1', conn)
```

```python
foul_info.head()
```

```python
foul_info['PeriodID'] = foul_info['Period'].map({'1st Q': 1
                                                , '2nd Q': 2
                                                , '3rd Q': 3
                                                , '4th Q': 4
                                                })
foul_info['NextPeriodID'] = foul_info['PeriodID'].map({1:2, 2:3, 3:4})
```

```python
foul_info.head()
```

```python
foul_info[foul_info['PeriodID'] < 4][['Period','RollingFoulCount','PeriodID']].groupby(['Period','RollingFoulCount'], as_index = False
                  ).count().rename(columns = {'PeriodID':'TotalPlayers'})
```
